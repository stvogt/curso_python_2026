## Ejercicio: construir conteos y graficar

Encuesta:

- Cada elemento es el transporte principal: `"bus"`, `"metro"`, `"auto"`, `"bici"`, `"a_pie"`

Tarea:

1) crea un diccionario `conteos` con el total por categoría usando un `for`
2) grafica con `plt.bar`
3) agrega título y ejes


## Solución

````python
# Datos de sociología: respuestas de una encuesta
respuestas = [
    "bus", "bus", "metro", "auto", "bus", "bici", "a_pie", "metro",
    "auto", "bus", "metro", "bici", "bus", "bus", "a_pie", "metro",
    "auto", "bus", "bici", "metro", "bus", "a_pie"
]

conteo = {}
for mt in respuestas:
    if not mt in conteo:
        conteo[mt] = 1
    else:
        conteo[mt] = conteo[mt] + 1

print(conteo)
plt.bar(conteo.keys(), conteo.values())
plt.title("Encuesta: Transporte principal")
plt.xlabel("Categoría")
plt.ylabel("Número de respuestas")
plt.grid(True)
plt.show()
# TU CÓDIGO
# 1) construir conteos = {...}
# 2) preparar listas categorias y valores
# 3) plt.bar(...)
# 4) title/xlabel/ylabel
# 5) show
````

## Solución

````python
# Solución
conteos = {}
for r in respuestas:
    if r in conteos:
        conteos[r] = conteos[r] + 1
    else:
        conteos[r] = 1

categorias = []
valores = []
for categoria in conteos:
    categorias.append(categoria)
    valores.append(conteos[categoria])

plt.bar(categorias, valores)
plt.title("Encuesta: Transporte principal")
plt.xlabel("Categoría")
plt.ylabel("Número de respuestas")
````

## Demo: dos curvas y una leyenda

Situación: reacción simple donde

- reactivo disminuye
- producto aumenta

Cuando hay más de una curva, necesitas:

- `label=...`
- `plt.legend()`


````python
# Química: concentraciones vs tiempo
tiempo_s = np.linspace(0, 50, 101)

# Modelo simple: A -> B (cinética de primer orden)
k = 0.06
concentracion_A = 1.0 * np.exp(-k * tiempo_s)
concentracion_B = 1.0 - concentracion_A
````

## Resultado

````python
plt.plot(tiempo_s, concentracion_A, label="Reactivo A")
plt.plot(tiempo_s, concentracion_B, label="Producto B")

plt.title("Reacción A → B: concentración vs tiempo")
plt.xlabel("Tiempo (s)")
plt.ylabel("Concentración (u.a.)")
plt.legend(loc="lower left")
````

## Subplots: varias vistas en una figura

A veces una sola gráfica no es suficiente.  
**Subplots** permiten mostrar **múltiples gráficos relacionados** en una misma figura.

### Idea clave

- Una **figura** puede contener varios **ejes**
- Cada eje es un gráfico independiente
- Útil para comparar patrones con los **mismos datos** o **mismas escalas**




## Esquema básico
```python
fig, ejes = plt.subplots(1, 2)   # 1 fila, 2 columnas

ejes[0].plot(x, y1)
ejes[0].set_title("Gráfico A")

ejes[1].plot(x, y2)
ejes[1].set_title("Gráfico B")

plt.show()
```

## Ejemplo

````python
# Crear subplots: 1 fila, 2 columnas
fig, ejes = plt.subplots(1, 2)

# Subplot 1: concentración vs tiempo
ejes[0].plot(tiempo_s, concentracion_A)
ejes[0].set_title("Concentración A")
ejes[0].set_xlabel("Tiempo (s)")
ejes[0].set_ylabel("Concentración (u.a.)")

# Subplot 2: temperatura vs tiempo
ejes[1].plot(tiempo_s, concentracion_B)
ejes[1].set_title("Concentracion B")
ejes[1].set_xlabel("Tiempo (s)")
ejes[1].set_ylabel("Concentración (u.a.")
````
