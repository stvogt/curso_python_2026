### Ejercicio 0.1  `grupo_edad(edad)`

Define una función `grupo_edad(edad)` que devuelva:

- `"menor"` si `edad < 18`
- `"adulto"` si `18 <= edad < 65`
- `"mayor"` si `edad >= 65`

Prueba con: `12`, `25`, `70`.


````python
# Tu código aquí
````

**Solución**

````python
def grupo_edad(edad):
    if edad < 18:
        return "menor"
    elif edad < 65:
        return "adulto"
    else:
        return "mayor"

for e in [12, 25, 70]:
    print(e, "->", grupo_edad(e))
````

### Ejercicio 0.3 

Define `contar_vocales(palabra)` que devuelva cuántas vocales hay en un texto.

Considera vocales: `a, e, i, o, u` (minúsculas). Usa `for` + `if`.

Prueba con: `"microbio"`, `"quimica"`.


````python
# Tu código aquí
````

**Solución**

````python
def contar_vocales(palabra):
    vocales = "aeiou"
    c = 0
    for ch in palabra:
        if ch in vocales:
            c = c + 1
    return c

print(contar_vocales("microbio"))
print(contar_vocales("quimica"))
````

### Ejercicio 1.1 — Microbiología: `total_colonias(conteos)`

En un experimento, `conteos` es una lista con el número de colonias por placa.

Implementa `total_colonias(conteos)` que devuelva la suma total.

Prueba con: `[12, 8, 15, 0]` (resultado esperado: `35`).


````python
# Tu código aquí
````

**Solución**

````python
def total_colonias(conteos):
    total = 0
    for x in conteos:
        total = total + x
    return total

print(total_colonias([12, 8, 15, 0]))
````

### Ejercicio 1.3 — Química: `max_concentracion(concs)`

`concs` es una lista de concentraciones (mM).

Implementa `max_concentracion(concs)` sin usar `max(...)`.

Prueba con `[0.2, 0.5, 0.1, 0.9]` (esperado: `0.9`).


````python
# Tu código aquí
````

**Solución**

````python
def max_concentracion(concs):
    m = concs[0]
    for x in concs:
        if x > m:
            m = x
    return m

print(max_concentracion([0.2, 0.5, 0.1, 0.9]))
````

### Ejercicio 1.4 — Sociología: `proporcion_si(respuestas)`

`respuestas` es una lista con textos `"si"` o `"no"`.

Implementa `proporcion_si(respuestas)` que devuelva la proporción de `"si"`:

- cuenta cuántos `"si"` hay
- divide por el total (`len(...)`)

Prueba con: `["si", "no", "si", "si"]` (esperado: `0.75`).


````python
# Tu código aquí
````

**Solución**

````python
def proporcion_si(respuestas):
    c = 0
    for r in respuestas:
        if r == "si":
            c = c + 1
    return c / len(respuestas)

print(proporcion_si(["si", "no", "si", "si"]))
````

### Ejercicio 2.1 — Lingüística: frecuencias de palabras

Dada una lista `palabras`, construye un diccionario `freq` donde:

- clave: palabra
- valor: cuántas veces aparece

Usa `for` + `if` y el patrón `if w in freq`.

Prueba con:
`palabras = ["microbio", "celula", "microbio", "gen", "celula"]`

Resultado esperado (orden no importa):
- `"microbio"` → `2`
- `"celula"` → `2`
- `"gen"` → `1`


````python
# Tu código aquí
````

**Solución**

````python
palabras = ["microbio", "celula", "microbio", "gen", "celula"]
freq = {}
for w in palabras:
    if w in freq:
        freq[w] = freq[w] + 1
    else:
        freq[w] = 1
print(freq)
````

### Ejercicio 3.1 — Química: normalizar composición (%)

Tienes un diccionario `mezcla` con masas (g) de cada componente.

Implementa `porcentajes(mezcla)` que devuelva un diccionario nuevo con porcentajes:
- suma total de masas
- porcentaje = (masa / total) * 100

Usa `for componente, masa in mezcla.items()`.

Prueba con:
```python
mezcla = {"NaCl": 2.0, "H2O": 18.0}
```


````python
# Tu código aquí
````

**Solución**

````python
def porcentajes(mezcla):
    total = 0
    for comp, masa in mezcla.items():
        total = total + masa
    out = {}
    for comp, masa in mezcla.items():
        out[comp] = (masa / total) * 100
    return out

mezcla = {"NaCl": 2.0, "H2O": 18.0}
print(porcentajes(mezcla))
````

### Ejercicio 3.2 

Implementa `longitudes(palabras)` que reciba una lista de palabras y devuelva un diccionario:

- clave: palabra
- valor: longitud (`len(palabra)`)

Prueba con: `["gen", "proteina", "microbio"]`.


````python
# Tu código aquí
````

**Solución**

````python
def longitudes(palabras):
    out = {}
    for w in palabras:
        out[w] = len(w)
    return out

print(longitudes(["gen", "proteina", "microbio"]))
````

### Ejercicio 3.3 

Tienes un diccionario `placas` con el conteo de colonias por placa:

```python
placas = {"P1": 12, "P2": 2, "P3": 25}
```

Implementa `clasificar_placas(placas, umbral)` que devuelva un diccionario nuevo:
- `"alta"` si conteo `>= umbral`
- `"baja"` si conteo `< umbral`

Prueba con `umbral = 10`.


````python
# Tu código aquí
````

**Solución**

````python
def clasificar_placas(placas, umbral):
    out = {}
    for placa, n in placas.items():
        if n >= umbral:
            out[placa] = "alta"
        else:
            out[placa] = "baja"
    return out

placas = {"P1": 12, "P2": 2, "P3": 25}
print(clasificar_placas(placas, 10))
````

### Ejercicio 5.1 — Microbiología: reporte por especie

Tienes conteos de colonias por especie:

```python
conteos = {
    "E_coli": 12,
    "S_aureus": 7,
    "B_subtilis": 5
}
```

Implementa `reporte_colonias(conteos, umbral)` que devuelva un diccionario nuevo:

- clave: especie
- valor: otro diccionario con:
  - `"n"`: conteo
  - `"categoria"`: `"alta"` si `n >= umbral`, si no `"baja"`

Luego imprime un reporte con `for especie, info in ...items():`.


````python
# Tu código aquí
````

**Solución**

````python
def reporte_colonias(conteos, umbral):
    rep = {}
    for especie, n in conteos.items():
        if n >= umbral:
            cat = "alta"
        else:
            cat = "baja"
        rep[especie] = {"n": n, "categoria": cat}
    return rep

conteos = {"E_coli": 12, "S_aureus": 7, "B_subtilis": 5}
rep = reporte_colonias(conteos, 8)
for especie, info in rep.items():
    print(f"{especie}: n={info['n']}, categoria={info['categoria']}")
````

### Ejercicio 5.2

Escribe la función `reporte_participacion(comunas, umbral)` que devuelva un
diccionario nuevo:

- clave: comuna  
- valor: `{"n": participantes, "categoria": "alta" o "baja"}`  

donde `"alta"` si `n >= umbral`, `"baja"` en otro caso.

Luego imprime un reporte con el formato:

Comuna: n=XX, categoria=YY



```python
comunas = {
    "Concepción": 120,
    "Chillán": 80,
    "Talcahuano": 95,
    "Los Ángeles": 60
}
```


**Solución**

````python
# Solución

comunas = {
    "Concepción": 120,
    "Chillán": 80,
    "Talcahuano": 95,
    "Los Ángeles": 60
}

def reporte_participacion(comunas, umbral):
    rep = {}
    for comuna, n in comunas.items():
        if n >= umbral:
            categoria = "alta"
        else:
            categoria = "baja"
        rep[comuna] = {"n": n, "categoria": categoria}
    return rep


rep = reporte_participacion(comunas, 90)

for comuna, info in rep.items():
    print(f"{comuna}: n={info['n']}, categoria={info['categoria']}")
````
