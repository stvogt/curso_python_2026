### Ejercicio 4.4 — Acumuladores múltiples

**Tiempo sugerido:** 8 min

Dada `nums = [3, -1, 0, 7, -5]`, calcula (1) la suma de los positivos estrictos y (2) la cantidad de negativos. Imprime ambas en una sola línea con f-string.

**Restricciones**
- Usa dos variables: `suma_pos` y `cant_neg`.
- Debes usar `for` y `if/elif/else` dentro.
- Imprime exactamente una línea.


```python
nums = [3, -1, 0, 7, -5]

# Tu código aquí
```

**Solución**

Nota: `pass` no es necesario; podrías omitir el `else` completo. Se incluye solo para dejar explícito el caso `0`.


```python
nums = [3, -1, 0, 7, -5]

suma_pos = 0
cant_neg = 0

for x in nums:
    if x > 0:
        suma_pos = suma_pos + x
    elif x < 0:
        cant_neg = cant_neg + 1
    else:
        # x == 0: no cambia nada
        cant_neg = cant_neg + 0

print(f"suma_pos={suma_pos}, cant_neg={cant_neg}")
```

# Sección 5 — Problema integrador

## Ejercicio 5.1 — Reporte de mediciones experimentales

**Tiempo sugerido:** 22 min

En un experimento se midió la **temperatura (°C)** de un sistema en distintos instantes.
Debes construir un reporte básico a partir de las mediciones.

Dada la lista `temperaturas`, calcula:

1) `n_total`: cantidad total de mediciones  
2) `promedio`: temperatura promedio (float)  
3) `en_rango`: cantidad de mediciones en el rango **[20.0, 25.0]** (incluye bordes)  
4) `fuera_rango`: cantidad de mediciones fuera de ese rango  
5) Clasificación de cada medición:
   - `Baja`: temperatura < 20.0  
   - `Normal`: 20.0 ≤ temperatura ≤ 25.0  
   - `Alta`: temperatura > 25.0  

Imprime **exactamente 3 líneas**:

- Línea 1: `Total: <n_total>`
- Línea 2: `Promedio: <promedio con 2 decimales>`
- Línea 3: `Baja:<baja> Normal:<normal> Alta:<alta> | En rango:<en_rango> Fuera:<fuera_rango>`

**Restricciones**
- No uses funciones ni imports.
- Usa un `for` sobre `temperaturas`.
- Debes usar acumuladores y contadores.
- Debes usar `if / elif / else` dentro del loop.
- El promedio debe calcularse con división en `float`.
- Imprime exactamente 3 líneas, como se indica.

**Pista (opcional):**
Inicializa todos los contadores en 0 antes del `for`. Dentro del loop actualiza suma, rangos y categorías.


**Solución**


```python
# Solución

temperaturas = [18.5, 21.0, 24.3, 26.1, 19.8, 22.5, 25.0, 27.4, 23.1]

n_total = len(temperaturas)

suma = 0.0
en_rango = 0
fuera_rango = 0

baja = 0
normal = 0
alta = 0

for t in temperaturas:
    suma = suma + t

    if (t >= 20.0) and (t <= 25.0):
        en_rango = en_rango + 1
    else:
        fuera_rango = fuera_rango + 1

    if t < 20.0:
        baja = baja + 1
    elif t <= 25.0:
        normal = normal + 1
    else:
        alta = alta + 1

promedio = suma / float(n_total)

print(f"Total: {n_total}")
print(f"Promedio: {promedio:.2f}")
print(f"Baja:{baja} Normal:{normal} Alta:{alta} | En rango:{en_rango} Fuera:{fuera_rango}")
```
